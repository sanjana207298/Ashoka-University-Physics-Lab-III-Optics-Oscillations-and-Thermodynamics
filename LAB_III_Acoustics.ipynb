```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit
import scipy.stats as stats
from scipy.stats import linregress
from tabulate import tabulate
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
```

```python
#L1
x = np.array([0.04273572815, 0.05856082398, 0.02028515727, 0.1000910681, 0.08934440909])
y = np.array([2, 3, 1, 5, 4])

m, c = np.polyfit(x, y, 1)
y2 = [m * i + c for i in x]

plt.scatter(x, y)
plt.errorbar(x, y, yerr=0, xerr=0.001, fmt='o', label='Data Point', markersize=4)
plt.plot(x, y2, 'r', label=f'Line of Best fit: p = {m:.4f}$\\sqrt{{(\\frac{{m}}{{T}})}}.2L$ + {c:.4f}')
plt.xlabel('$\\sqrt{{(\\frac{{m}}{{T}})}}.2L$ (s)')
plt.ylabel('No. of loops (p)')
plt.legend()
plt.show()

slope, intercept, r_value, p_value, std_err = linregress(x, y)
slope_error = std_err / np.sqrt(len(x))

print(f"Slope: {slope:.4f}")
print(f"Slope error: {slope_error:.4f}")
print(2*(50-slope))
```
