```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
from scipy.signal import find_peaks
import pandas as pd
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
```

#PART A

```python
#Spring 1 has an orange sticker
#Spring 2 has an orange sticker and is marked
#Spring 3 has no sticker

#Unstretched Length
s1x0 = 12.0/100 #m
s2x0 = 12.3/100 #m
s3x0 = 12.1/100 #m

#Stretched Length under the influence of only gravity and no masses
s1x1 = (41.0-26.4)/100 #m
s2x1 = (41.2-26.9)/100 #m
s3x1 = (40.9-28.5)/100 #m

#Mass
s1m = 20.8/1000 #kg
s2m = 21.2/1000 #kg
s3m = 21.3/1000 #kg
ms = np.array([s1m,s2m,s3m])
```

Static

```python
#Spring 1
exts1 = np.array([41-26.8,41-25,41-23.3,41-19.3,40.9-15.2])/100 - s1x0
#Spring 2
exts2 = np.array([41.1-27.6,41-26.3,41-24.1,41-19.3,41-14])/100 - s2x0
#Spring 3
exts3 = np.array([40.8-28.3,40.9-27.8,41.1-26.5,41.1-21.7,41.1-16.1])/100 - s3x0

masses = np.array([1.3,5.6,10.0,20.3,30.3])/1000

plt.scatter(masses, exts1)
m1,c1 = np.polyfit(masses, exts1,1)
X1 =[ m1*i + c1 for i in masses ]
print(linregress(masses,exts1))
m_err= 0.0001
x_err= 0.001
plt.errorbar(masses,exts1, xerr=m_err, yerr=x_err, fmt='o', label = 'Data Point', markersize = 4)
plt.plot(masses, X1, 'r', label='Line of best fit: x = {}m + {}'.format(round(m1, 4), round(c1, 4)))
plt.xlabel('Attached mass (kg)')
plt.ylabel('Extension (m)')
plt.legend()
plt.show()
g = 9.79 #value in Sonepat
k1 = g/m1
print('slope =', m1, 'intercept=', c1,"k=", k1)

plt.scatter(masses, exts2)
m2,c2 = np.polyfit(masses, exts2,1)
X2 =[ m2*i + c2 for i in masses ]
print(linregress(masses,exts2))
plt.plot(masses, X2, 'r', label='Line of best fit: x = {}m + {}'.format(round(m2, 4), round(c2, 4)))
plt.errorbar(masses,exts2, xerr=m_err, yerr=x_err, fmt='o', label = 'Data Point', markersize = 4)
plt.xlabel('Attached mass (kg)')
plt.ylabel('Extension (m)')
plt.legend()
plt.show()
k2 = g/m2
print('slope =', m2, 'intercept=', c2,"k=", k2)


plt.scatter(masses, exts3)
m3,c3 = np.polyfit(masses, exts3,1)
X3 =[ m3*i + c3 for i in masses ]
print(linregress(masses,exts3))
plt.plot(masses, X3, 'r', label='Line of best fit: x = {}m  {}'.format(round(m3, 4), round(c3, 4)))
plt.errorbar(masses,exts3, xerr=m_err, yerr=x_err, fmt='o', label = 'Data Point', markersize = 4)
plt.xlabel('Attached mass (kg)')
plt.ylabel('Extension (m)')
plt.legend()
plt.show()
k3 = g/m3
print('slope =', m3, 'intercept=', c3,"k=", k3)
print(linregress(masses,exts3))

print('EXTENSIONS=', exts1, exts2, exts3)
```

Dynamic

```python
times1 = np.array([7.54,8.91,10.28,12.98,15.12])/19.5
times2 = np.array([7.47,9.52,11.45,14.35,16.82])/19.5
times3 = np.array([6.98,9.47,11.09,14.67,17.22])/19.5
tsq3, tsq1, tsq2 = times3**2, times1**2, times2**2

plt.scatter(masses, tsq1)
m1,c1 = np.polyfit(masses, tsq1,1)
T1 =[ m1*i + c1 for i in masses ]
print(linregress(masses,tsq1))
m_err= 0.0001
T_err= 0.01
plt.errorbar(masses,tsq1, xerr=m_err, yerr=T_err, fmt='o', label = 'Data Point', markersize = 4)
plt.plot(masses, T1, 'r', label='Line of best fit: $T^2$ = {}m + {}'.format(round(m1, 4), round(c1, 4)))
plt.xlabel('Attached mass (kg)')
plt.ylabel('Time Period of Oscillation (s)')
plt.legend()
plt.show()
k1 = (4*np.pi**2)/m1
print('slope =', m1, 'intercept=', c1,"k=", k1)

plt.scatter(masses, tsq2)
m2,c2 = np.polyfit(masses, tsq2,1)
T2 =[ m2*i + c2 for i in masses ]
print(linregress(masses,tsq2))
m_err= 0.0001
T_err= 0.01
plt.errorbar(masses,tsq2, xerr=m_err, yerr=T_err, fmt='o', label = 'Data Point', markersize = 4)
plt.plot(masses, T2, 'r', label='Line of best fit: $T^2$ = {}m + {}'.format(round(m2, 4), round(c2, 4)))
plt.xlabel('Attached mass (kg)')
plt.ylabel('Time Period of Oscillation (s)')
plt.legend()
plt.show()
k2 = (4*np.pi**2)/m2
print('slope =', m2, 'intercept=', c2,"k=", k2)

plt.scatter(masses, tsq3)
m3,c3 = np.polyfit(masses, tsq3,1)
T3 =[ m3*i + c3 for i in masses ]
print(linregress(masses,tsq3))
m_err= 0.0001
T_err= 0.01
plt.errorbar(masses,tsq3, xerr=m_err, yerr=T_err, fmt='o', label = 'Data Point', markersize = 4)
plt.plot(masses, T3, 'r', label='Line of best fit: $T^2$ = {}m + {}'.format(round(m3, 4), round(c3, 4)))
plt.xlabel('Attached mass (kg)')
plt.ylabel('Time Period of Oscillation (s)')
plt.legend()
plt.show()
k3 = (4*np.pi**2)/m3
print('slope =', m3, 'intercept=', c3,"k=", k3)
print('TIME PERIOD SQUARES=', tsq1, tsq2, tsq3)
```

```python
def normalize(arr):
    arr_min = arr.min()
    arr_max = arr.max()
    return (arr - arr_min) / (arr_max - arr_min)

def analyse(left_bob_path, right_bob_path):
  #t_L, x_L = np.loadtxt(left_bob_path, delimiter=",", unpack=True, skiprows=2, usecols=(0,1))
  #t_R, x_R = np.loadtxt(right_bob_path, delimiter=",", unpack=True, skiprows=2, usecols=(0,1))

  left_bob_data = pd.read_csv(left_bob_path, skiprows=2, usecols=[0, 1], header=None, names=['t_L', 'x_L'])
  right_bob_data = pd.read_csv(right_bob_path, skiprows=2, usecols=[0, 1], header=None, names=['t_R', 'x_R'])

  # Extract data from DataFrames
  t_L = left_bob_data['t_L'].values
  x_L = left_bob_data['x_L'].values
  t_R = right_bob_data['t_R'].values
  x_R = right_bob_data['x_R'].values

  x_L_norm = normalize(x_L)
  x_R_norm = normalize(x_R)

  fft_L = np.fft.fft(x_L_norm)
  fft_R = np.fft.fft(x_R_norm)
  freq_L = np.fft.fftfreq(len(x_L_norm), t_L[1]-t_L[0])
  freq_R = np.fft.fftfreq(len(x_R_norm), t_R[1]-t_R[0])
  peaks_L = np.argsort(np.abs(fft_L))[::-1][:5] #highest 5 peaks
  peak_freqs_L = freq_L[peaks_L]
  peaks_R = np.argsort(np.abs(fft_R))[::-1][:5] #highest 5 peaks
  peak_freqs_R = freq_R[peaks_R]

  peaks_L,_ = find_peaks(np.abs(fft_L))
  peaks_R,_ = find_peaks(np.abs(fft_R))

  fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 8))
  fig.subplots_adjust(wspace=0.15, hspace=0.3)

  axes[0,0].plot(t_L, x_L_norm, label='Left mass')
  axes[0,0].plot(t_R, x_R_norm, label='Right mass')
  #axes[0,0].set_xlim(0,20)
  axes[0,0].set_xlabel('Time (s)')
  axes[0,0].set_ylabel('Distance (cm)')
  axes[0,0].set_title('Position-Time')

  axes[0,1].plot(t_L, x_L_norm)
  axes[0,1].plot(t_R, x_R_norm)
  axes[0,1].set_xlim(10,20)
  axes[0,1].set_xlabel('Time (s)')
  axes[0,1].set_ylabel('Distance (cm)')
  axes[0,1].set_title('Position-Time')

  axes[1,0].plot(freq_L, np.abs(fft_L))
  axes[1,0].plot(freq_R, np.abs(fft_R))
  #axes[1,0].set_xlim(0.5, 1)
  #axes[1,0].set_ylim(0,500)
  axes[1,0].set_xlabel('Frequency (Hz)')
  axes[1,0].set_ylabel('Magnitude')
  axes[1,0].set_title('Fourier Transform')

  axes[1,1].plot(freq_L, np.abs(fft_L))
  axes[1,1].plot(freq_R, np.abs(fft_R))
  axes[1,1].scatter(freq_L[peaks_L],np.abs(fft_L)[peaks_L] , color='g')
  axes[1,1].scatter(freq_R[peaks_R],np.abs(fft_R)[peaks_R] , color='r')
  axes[1,1].set_xlim(0.3, 1.3)
  axes[1,1].set_ylim(0,500)
  axes[1,1].set_xlabel('Frequency $\omega$ (Hz)')
  axes[1,1].set_ylabel('FFT Amplitude')
  axes[1,1].set_title('Fourier Transform')

  handles, labels = [], []
  for ax in fig.axes:
      h, l = ax.get_legend_handles_labels()
      handles += h
      labels += l
  fig.legend(handles, labels, loc='lower center')

  plt.show()
  print("Peak frequencies of left mass:", peak_freqs_L)
  print("Peak frequencies of right mass:", peak_freqs_R)
```
